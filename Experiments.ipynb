```python
n_runs = 50
M = 8
n_pert= 200
sigma = 1
alpha = 0.01
n_episodes = 1
iterations = 200

df_runs = pd.DataFrame({'Time': range(iterations)})

# Print the first few rows of the DataFrame
# print(df.head())
for i_run in range(n_runs):
    history, theta_t = Simulation(M = M, n_pert=n_pert, sigma=sigma, alpha=alpha, n_episodes=n_episodes, iterations=iterations)
    hist_size = len(history)
    if(hist_size<iterations):
        for i in range(hist_size, iterations):
            history.append((i, 500.0))
    Js = [pair[1] for pair in history]
    df_runs[f"hist_{i_run}"] = Js

df_melt = df_runs.melt(id_vars="Time", var_name="hist_i", value_name="J")

sns.lineplot(df_melt, x='Time', y='J', err_style="band", errorbar=("sd", 1))

plt.ylabel(f"J-estimate Average over {n_runs} runs")
plt.xlabel(f"Iterations/Time of 1 ES algorithm")
plt.title(f"Learning Curve\
            \nMean J-estimate vs Iterations with 1 Standard Deviation interval for {iterations} steps for {n_runs} ES runs\
            \nM = {M}, n_pert = {n_pert}, sigma = {sigma}, alpha = {alpha}")

plt.savefig(f"plots/nruns = {n_runs} M = {M}, n_pert = {n_pert}, sigma = {sigma}, alpha = {alpha}.jpg", bbox_inches='tight')
plt.show()
```
